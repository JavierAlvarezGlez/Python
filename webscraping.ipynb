```python
import requests
```

```python
url = "https://elpais.com"
ret = requests.get(url)
```

```python
ret.status_code
```

```python
html = ret.text  # html de EL Pais
```

```python
soup = BeautifulSoup(html, "html.parser")
```

```python
[a["href"] for a in soup.find_all("a")] # ancors
# Esto sirve para coger los hiperenlaces de cada noticia del periodico
```

```python
soup.find_all("img") # Es para buscar la palabra clave, pero solo funciona con imagenes
```
